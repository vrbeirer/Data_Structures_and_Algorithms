```python
def print_factors(m, n):
    fm = []  
    for i in range(1, m + 1):
        if m % i == 0:
            fm.append(i)  

    fn = []  
    for j in range(1, n + 1):
        if n % j == 0:
            fn.append(j)  

    cf = []
    for f in fm:
        if f in fn:
            cf.append(f)  

    return cf[-1]  


result = print_factors(14,63)
print(result)
```

```python
def gcd(m,n):
    cf = []
    min1 = min(m,n)
    for i in range(1,min1 + 1):
        if m%i == 0 and n%i == 0:
            cf.append(i)
    return cf[-1]

gcd(14,63)
```

```python
def gcd1(m,n):
    i = min(m,n)
    while i > 0:
        if m%i == 0 and n%i == 0:
            return i
        else:
            i = i-1

gcd1(14,63)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define range for m and n (assuming m = n for simplicity in visualization)
m_values = np.linspace(1, 100, 100)  # Values from 1 to 100

# Compute time complexity values
time_on2 = m_values ** 2  # O(m * n), assuming m = n
time_log = np.log2(m_values)  # O(log(m))

# Plot the graph
plt.figure(figsize=(8, 5))
plt.plot(m_values, time_on2, label="O(m * n) (Current Approach)", color='r')
plt.plot(m_values, time_log, label="O(log(m)) (Optimized GCD)", color='b')

# Labels and title
plt.xlabel("Value of m (assuming m = n)")
plt.ylabel("Time Complexity (Arbitrary Units)")
plt.title("Time Complexity Comparison: O(m * n) vs O(log m)")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
