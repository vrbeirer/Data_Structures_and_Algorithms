```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DLL:
    def __init__(self):
        self.head = None

    # Insert at the beginning of the list
    def insert_beg(self, data):
        x = Node(data)
        if self.head is None:  # If the list is empty
            self.head = x
        else:
            x.next = self.head
            self.head.prev = x
            self.head = x

    # Insert at the end of the list
    def insert_end(self, data):
        y = Node(data)
        if self.head is None:  # If the list is empty
            self.head = y
            return
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = y
        y.prev = temp

    # Insert at a specific position in the list
    def insert_pos(self, data, pos):
        if pos < 1:
            print("Position should be >= 1")
            return

        n = Node(data)
        if pos == 1:  # Insert at the beginning
            self.insert_beg(data)
            return

        temp = self.head
        for i in range(1, pos - 1):
            if temp is None:
                print("Position out of bounds")
                return
            temp = temp.next
        
        # Insert at the end
        if temp.next is None:
            self.insert_end(data)
            return

        n.next = temp.next
        n.prev = temp
        temp.next.prev = n
        temp.next = n

    # Display the list
    def display(self):
        if self.head is None:
            print("LL is Empty")
        else:
            temp = self.head
            while temp:
                print(temp.data, "-->", end=" ")
                temp = temp.next
            print("None")

# Example Usage
D = DLL()

# Adding nodes manually to the list
n1 = Node(10)
D.head = n1
n2 = Node(20)
n2.prev = n1
n1.next = n2
n3 = Node(30)
n3.prev = n2
n2.next = n3

# Inserting nodes at different positions
D.insert_beg(5)  # Insert at the beginning
D.insert_end(40) # Insert at the end
D.insert_pos(25, 4)  # Insert at position 4

D.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DLL:
    def __init__(self):
        self.head = None

    def delete_beg(self):
        if self.head:
            self.head = self.head.next
            if self.head:
                self.head.prev = None

    def delete_end(self):
        if self.head:
            temp = self.head
            while temp.next:
                temp = temp.next
            if temp.prev:
                temp.prev.next = None
            else:
                self.head = None

    def delete_pos(self, pos):
        if self.head:
            temp = self.head
            if pos == 1:
                self.delete_beg()
                return
            for _ in range(1, pos):
                if temp.next is None:
                    print("Position out of range")
                    return
                temp = temp.next
            temp.prev.next = temp.next
            if temp.next:
                temp.next.prev = temp.prev

    def display(self):
        if not self.head:
            print("List is Empty")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" --> ")
            temp = temp.next
        print("None")

# Example Usage
D = DLL()
D.head = Node(10)
D.head.next = Node(20)
D.head.next.prev = D.head
D.head.next.next = Node(30)
D.head.next.next.prev = D.head.next

D.delete_beg()
D.delete_end()
D.delete_pos(1)
D.display()
```
